# Week 12: Solution to Meeting at a Central Point

Please help the caterpillars find a vertex where they can meet soon!

Three caterpillars are sitting on vertices $v_1$, $v_2$, and $v_3$ of a connected graph $G = (V,E)$. The three caterpillars would like to meet at one of the vertices of the graph. All they can do is move from one vertex to another via the edges of the graph, and they can only stop at the vertices of the graph, not on the edges. When traversing an edge, the three caterpillars travel at the same speed of one edge per hour. For an example, consider the graph below. The starting points of the caterpillars are marked in red.

```python
import networkx as nx

# Create graph G
G = nx.Graph()
G.add_nodes_from(['v1','v2','v3','v4','v5','v6','v7','v8','v9'])
G.add_edges_from([('v1','v4'), ('v1','v6'),
                  ('v2','v5'), ('v2','v4'),
                  ('v3','v8'), ('v3','v9'),
                  ('v4','v6'), ('v4','v8'), ('v4','v9'),
                  ('v5','v9'),
                  ('v6','v8'),
                  ('v7','v8'),
                  ('v8','v9')])

# Plot graph G
%matplotlib inline
nx.draw(G, with_labels=True, node_color=3*['lightcoral']+6*['lightblue'])
```

To meet at vertex $v_5$, the caterpillar from vertex $v_1$ could travel to $v_5$ via the vertices $v_4$ and $v_2$ in three hours (and there is no faster option). The caterpillars from $v_2$ and $v_3$ could be there in one and two hours, respectively. Thus, the earliest meeting at $v_5$ could take place three hours after the caterpillars start moving.

### Your Task

The caterpillars want to know at which vertex they can meet as soon as possible. Implement an algorithm to determine such a vertex. You can either follow the steps we provided below or implement your own algorithm.

*Hint: You may use the implementation of Breadth-first search `nx.bfs_edges(G, source)`.*

```python
# STEP 1: Write a function that computes the distances of all nodes from a source node
# Input:  - a graph object 'G' representing the given graph
#         - a string 'source' representing the source node 
# Output: - a dictionary, with node names being keys and distances to the source node being values

def distances(G, source):
    edges = list(nx.bfs_edges(G, source))      # apply breadth first search
    dist = {node: 0 for node in list(G.nodes)} # initialize all distances as zero
    for edge in edges:
        dist[edge[1]] = dist[edge[0]] + 1
    return dist

print(distances(G, 'v1'))
```

```python
# STEP 2: Write a function that computes the maximum time needed for all caterpillars to reach a vertex
# Input:  - a graph object 'G' representing the given graph
#         - a list of strings 'sources' representing the source nodes of caterpillars 
# Output: - a dictionary, with node names being keys and time needed to reach the node being values

def time_to_meeting(G, sources):
    time = {node: 0 for node in list(G.nodes)} # initialize all distances as zero
    dist = [distances(G, source) for source in sources]
    for node in list(G.nodes):
        time[node] = max([d[node] for d in dist])
    return time

print(time_to_meeting(G, ['v1','v2','v3']))
```

```python
# STEP 3: Apply your function to the graph G and get a possible meeting vertex

time = time_to_meeting(G, ['v1','v2','v3'])

print(f'A possible meeting point is the vertex {min(time, key=time.get)}.') 
```

Analyze the running time of your algorithm in terms of $|V|$ and $|E|$. Note that the number of caterpillars is assumed to be constant. What is the complexity of your algorithm? Can you achieve a complexity of $O(|V|+|E|)$?

<font color='blue'>

**Solution:** Remember that BFS takes $O(|V|+|E|)$ time.

1. We apply BFS three times, which has a complexity of $O(|V|+|E|)$.
2. We compute the maximum of three values $|V|$ times, which is of complexity $O(|V|)$.
3. We find the optimal vertex by comparing $|V|$ values, which is again of complexity $O(|V|)$.

Hence, the complete procedure is of complexity $O(|V|+|E|)$.
